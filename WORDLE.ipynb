```python
import pandas as pd
```

```python
words = pd.read_csv(r'./Desktop/fiveletter.csv')

words.head()
```

```python
letter_index = {0: 'first letter',
                1: 'second letter',
                2: 'third letter',
                3: 'fourth letter',
                4: 'fifth letter'}
```

```python
for key in letter_index.keys():
    words[letter_index[key]] = words['word'].str[key]

# words['first letter'] = words['word'].str[0]
# words['second letter'] = words['word'].str[1]
# words['third letter'] = words['word'].str[2]
# words['fourth letter'] = words['word'].str[3]
# words['fifth letter'] = words['word'].str[4]

words.head()
```

```python
letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m',
           'n','o','p','q','r','s','t','u','v','w','x','y','z']

letter_counts = {letter: None for letter in letters}
```

```python
for letter in letters:
    count = 0
    for key in letter_index.keys():
        count += len(words[words[letter_index[key]] == letter])
    # count = len(words[words['first letter'] == letter]) + \
    #         len(words[words['second letter'] == letter]) + \
    #         len(words[words['third letter'] == letter]) + \
    #         len(words[words['fourth letter'] == letter]) + \
    #         len(words[words['fifth letter'] == letter])
    
    letter_counts[letter] = count
    
letter_counts
```

```python
for i in range(len(words)):
    word = str(words.loc[i, 'word'])
    count = 0
    for letter in list(set(word)):
        count += letter_counts[letter]
        
    words.loc[i,'count'] = count

words.head()
```

```python
words.sort_values('count', ascending=False, inplace=True, ignore_index=True)

words.head()
```

```python
def black_letter(letter: str):
    words.drop(words[(words['first letter'] == letter) | 
                    (words['second letter'] == letter) | 
                    (words['third letter'] == letter) | 
                    (words['fourth letter'] == letter) |
                    (words['fifth letter'] == letter)].index, 
               inplace=True)
```

```python
def yellow_letter(letter:str, column:int):
    words.drop(words[words[letter_index[column]] == letter].index, inplace=True)
```

```python
def green_letter(letter:str, column:int):
    words.drop(words[words[letter_index[column]] != letter].index, inplace=True)
```

```python
black_letter('a')
black_letter('r')
black_letter('o')
black_letter('s')
black_letter('e')

words.head()
```

```python
black_letter('u')
yellow_letter('n',1)
yellow_letter('l',2)
yellow_letter('i',3)
black_letter('t')

words.head()
```

```python
black_letter('b')
green_letter('l',1)
green_letter('i',2)
green_letter('n',3)
black_letter('d')

words.head()
```

```python
black_letter('c')
green_letter('g',4)

words.head()
```
